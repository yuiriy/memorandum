# 折れ線グラフ

```python
%matplotlib inline
import matplotlib.pyplot as plt

price = [100, 250, 380, 500, 700]
number = [1, 2, 3, 4, 5]

# グラフを書く
#marker:点の形を指定(x:×,o:●,s:■,^:▲)
#color:線の色
#linewidth:線の太さ(数字が大きいほど太い)
#linestyle:線のスタイル("dashed":点線)
#label:線の名前(凡例をつけるときに必要)
#色と点線は3つ目の引数にすることで短縮可能("r--":赤点線,"b--":青点線,"g--":緑点線)
#plt.plot(price, number, "r--", marker="x", linewidth=4)
plt.plot(price, number, marker="x", color="red", linewidth=2, linestyle = "dashed", label = "a")

# グラフのタイトル
plt.title("price / number")

# x軸のラベル
plt.xlabel("price")

# y軸のラベル
plt.ylabel("number")

# グリッド（縦横線）を表示する
#axis='y':横軸のみ
#axis='x':縦軸のみ
plt.grid(axis = 'y')

#グラフの凡例
#loc:凡例の位置(1:右上,2:左上,3:左下,4:右下)
plt.legend(loc=2)

#画像として保存する
plt.savefig("linegraph.png")

# 表示する
plt.show()
```

# 複数の折れ線グラフ

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-10,10,100)
y = x**2

#グラフを書く
#marker:点の形を指定(x:×,o:●,s:■,^:▲)
#color:線の色
#linewidth:線の太さ(数字が大きいほど太い)
#linestyle:線のスタイル("dashed":点線)
#label:線の名前(凡例をつけるときに必要)
#色と点線は3つ目の引数にすることで短縮可能("r--":赤点線,"b--":青点線,"g--":緑点線)
#plt.plot(price, number, "r--", marker="x", linewidth=4)
plt.plot(x/2,y/2, linestyle="dashed", color = "red", label = "Dashed")
plt.plot(x, y, label = "Line")

# グラフのタイトル
plt.title("title")

# x軸のラベル
plt.xlabel("x")

# y軸のラベル
plt.ylabel("y")

# グリッド（縦横線）を表示する
#axis='y':横軸のみ
#axis='x':縦軸のみ
plt.grid(axis='y')

#グラフの凡例
#loc:凡例の位置(1:右上,2:左上,3:左下,4:右下)
plt.legend(loc=4)

#画像として保存する
plt.savefig("graph.png")

#グラフを表示する
plt.show()
```

# 散布図

# 棒グラフ

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

labels = ["A", "B", "C", "D", "E", "F"]
x = range(0, 6)
y = [10, 20, 30, 40, 50, 60]
c = ["gray","blue", "gray", "gray", "gray", "gray"]
yerr = np.array([5, 10, 10, 10, 20, 20])

# 棒グラフを描画する
#color:線の色
#width:幅
#linewidth=0:枠線非表示
#yerr:y軸のエラーバー表示
#ecolor:エラーバーの色
#log=true:y軸を対数目盛で表示
plt.bar(x, y, tick_label = labels, color = c, yerr = yerr, ecolor = "red") 

# グラフのタイトル
plt.title("title")

# x軸のラベル
plt.xlabel("x")

# y軸のラベル
plt.ylabel("y")

# グリッド（縦横線）を表示する
#axis='y':横軸のみ
#axis='x':縦軸のみ
plt.grid()

#画像として保存する
plt.savefig("graph.png")

#グラフを表示する
plt.show()
```

# 積み上げ棒グラフ

```python
%matplotlib inline
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
height = [10, 20, 30, 40]
height2 =[5, 10, 15, 20]
labels = ["A", "B", "C", "D"]
 
# 棒グラフを描画する
#color:線の色
#width:幅
#linewidth=0:枠線非表示
#log=True:y軸を対数目盛で表示
plt.bar(x, height, tick_label = labels, label="blue")
plt.bar(x, height2, tick_label = labels, bottom=height, label = "orange")
 
# グラフのタイトル
plt.title("title")

# x軸のラベル
plt.xlabel("x")

# y軸のラベル
plt.ylabel("y")

# グリッド（縦横線）を表示する
#axis='y':横軸のみ
#axis='x':縦軸のみ
plt.grid()

#グラフの凡例
#loc:凡例の位置(1:右上,2:左上,3:左下,4:右下)
plt.legend(loc=2)

#画像として保存する
plt.savefig("graph.png")

#グラフを表示する
plt.show()
```

# 二系列の棒グラフ

# 複数系列の棒グラフ

```python
%matplotlib inline
import matplotlib.pyplot as plt

# 棒の配置位置、ラベルを用意
x = np.array([1, 2, 3, 4])
labels = ["Apple", "Banana", "Carrot", "Daikon"]
 
# 各系列のデータを用意
height = np.random.rand(4)
height2 = np.random.rand(4)
height3 = np.random.rand(4)
height4 = np.random.rand(4)
data = [height, height2,height3,height4]
 
# マージンを設定
margin = 0.2  #0 <margin< 1
totoal_width = 1 - margin
 
# 棒グラフをプロット
for i, h in enumerate(data):
  pos = x - totoal_width *( 1- (2*i+1)/len(data) )/2
  plt.bar(pos, h, width = totoal_width/len(data))
 
# ラベルの設定
plt.xticks(x, labels)
 
plt.show()
```

# ヒストグラム

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

#期待値0、標準偏差1の標準正規分布からランダムに1000個サンプルを摘出する
x = np.random.normal(0,1,1000)

#ヒストグラムを描写する
#bins:ビンの数
#color:線の色
#orientation="horizontal":横棒で描写
plt.hist(x, bins = 16)#, orientation="horizontal")

# グラフのタイトル
plt.title("title")

# x軸のラベル
plt.xlabel("x")

# y軸のラベル
plt.ylabel("y")

# グリッド（縦横線）を表示する
#axis='y':横軸のみ
#axis='x':縦軸のみ
plt.grid()

#画像として保存する
plt.savefig("graph.png")

#グラフを表示する
plt.show()
```

# 円グラフ

```python
%matplotlib inline
import matplotlib.pyplot as plt

# ラベルと値は反時計回りに指定する
labels = ["E", "D", "C", "B", "A"] 
y = [10, 20, 30, 40, 50] 

# 円グラフからの飛び出し
ex = [0, 0, 0, 0, 0.1] 

#なくてもよい（おまかせになる）
c = ["red", "lightcoral", "green", "yellow", "blue"]

# 円グラフを描く
#autopct:割合を表示する('%1.1f%%':少数第1位まで,'%d%%':整数)
#startangle:開始を12時にする
#counterclock:並び順(False:定義した順)
plt.pie(y, explode = ex, labels = labels, colors = c, autopct = '%1.1f%%', startangle = 90) 
#plt.pie(y, explode = ex, labels = labels, colors = c, autopct = '%1.1f%%', startangle = 90, counterclock=False)

# グラフのタイトル
plt.title('title')

#きれいな円グラフにする
plt.axis("equal")

#画像として保存する
plt.savefig("graph.png")

#グラフを表示する
plt.show()
```

# 複数のグラフを描画する

```python
%matplotlib inline
import matplotlib.pyplot as plt
fig = plt.figure()

# 1行2列の左
labels = ["A", "B", "C", "D", "E", "F"]
x = range(0, 6)
y = [10, 20, 30, 40, 50, 60]
ax1 = fig.add_subplot(1, 2, 1)#(何行目,何列目,並べる順番)
ax1.bar(x, y, tick_label = labels) 
ax1.set_title("graph1")

# 1行2列の右
labels = ["E", "D", "C", "B", "A"] 
y = [10, 20, 30, 40, 50] 
ex = [0, 0, 0, 0, 0.1] 
ax2 = fig.add_subplot(1, 2, 2)
ax2.pie(y, explode = ex, labels = labels, autopct = '%1.1f%%', startangle = 90) 
ax2.set_title("graph2")

#画像として保存する
plt.savefig("graph.png")

# グラフを表示する
plt.show()
```
